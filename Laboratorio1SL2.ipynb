## **Laboratorio 1**

### Carga de Librerías

```python
import numpy as np
import pandas as pd
from sklearn.metrics import roc_auc_score, roc_curve
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense
```

### Carga de los Datos ya tratados con Feature Engineering

```python
data_path1 = '/content/drive/MyDrive/DataTratadaTrain.csv'
data_path2 = '/content/drive/MyDrive/DataTratadaTest.csv'
datatrain = pd.read_csv(data_path1)
datatest = pd.read_csv(data_path2)
```

### Definición de Valores den entrenamiento y de validación

```python
X_train = datatrain.iloc[:, datatrain.columns != 'Survived']
y_train = datatrain.loc[:, 'Survived']
X_test = datatest.iloc[:, datatest.columns != 'Survived']
```

### Primera Estructura - Posición Kaggle: 14682

```python
inputs = X_train.shape[1]

#base secuencial para definir la estructura de la Red (SIEMPRE INICIAR CON SQUENTIAL)
clasificador = Sequential()

#primera capa oclta
clasificador.add(Dense(input_dim = inputs, units=15, activation='relu', kernel_initializer='uniform'))

#segunda capa oculta
#clasificador.add(Dense(units=6, activation='relu', kernel_initializer='uniform'))

#capa de salida
clasificador.add(Dense(units=1, activation='sigmoid', kernel_initializer='uniform'))
```

```python
#parametros de optimización
clasificador.compile(optimizer='SGD', loss='binary_crossentropy', metrics=['accuracy'])

#parametros de entrenamiento.
clasificador.fit(X_train, y_train, batch_size=25, epochs=550)
```

```python
# Guardardamos la red.
path = '/content/drive/MyDrive/' 
clasificador.save(path + '6_6_sgd_sig_out.h5')
```

```python
# Cargamos la red.
new_model = keras.models.load_model(path + '6_6_sgd_sig_out.h5')
```

```python
#predicciones
y_preds = new_model.predict(X_test)
preds1 = pd.DataFrame(y_preds, columns = ['Survived'])
preds1.loc[preds1.Survived>50]='1'
preds1.loc[preds1.Survived<50]='0'
preds1.insert(0,'PassengerId',datatest['PassengerId'])
preds1.to_csv("Pred1.csv")
```

### Métricas Estructura 1

```python
#Accuracy
y_preds = new_model.predict(X_train)
y_preds = (y_preds >= 0.5)
accuracy_score(y_train, y_preds)
```

```python
# ROC-AUC
roc_auc_score(y_train, y_preds, multi_class='ovr')
```

### Estructura Ocho - Posición Kaggle: 14682

```python
#KFolds
def red_v5():
  #inputs = X_train.shape[1]

  #base secuencial para definir la estructura de la Red (SIEMPRE INICIAR CON SQUENTIAL)
  clasificador = Sequential()

  #primera capa oclta
  clasificador.add(Dense(input_dim = 8, units=10, activation='sigmoid', kernel_initializer='uniform'))

  #segunda capa oculta
  clasificador.add(Dense(units=10, activation='sigmoid', kernel_initializer='uniform'))

  #capa de salida
  clasificador.add(Dense(units=1, activation='sigmoid', kernel_initializer='uniform'))

  #parametros de optimización
  clasificador.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
  
  return clasificador
```

```python
clasificador = KerasClassifier(build_fn=red_v5, batch_size=30, nb_epoch=1200)
```

```python
kfolds = KFold(n_splits=10, shuffle=True)
accs = cross_val_score(clasificador, X=X_train, y=y_train, cv=kfolds, verbose=1, n_jobs=-1)
```

```python
accs
```

### Conclusión

La mejor estructura es la 2 con 1 sola capa y 10 neuronas, optimizador RMSprop y activador sigmoide. Es la mejor ya que es la que logró el accuracy y el ROCAUC más altos, 0.79 y 0.76 respectivamente.
